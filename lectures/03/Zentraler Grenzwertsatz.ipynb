# Initialize

## Load Libraries

```python
import matplotlib
import random
import numpy as np
import matplotlib.pyplot as plt
```

# Main

## Generate Data - One Die

```python
# Generate a uniformly distributed random number, equivalent to throwing a die
d = np.random.randint(1, 7, 1)

# We use the numpy-function np.random.randint(low, high, size), which generates an array of size 'size' 
# which holds random integers drawn between [low, high[
```

```python
print(d)
```

## Plot Histogram

```python
# Number of bins and bin size
n_bins = 6
bin_size = 1

# Define an array of the bin edges. This we will hand over to the histogram function, the data will be sorted into bins, defined by these edges.
bins = np.linspace(1, 6 + 1, n_bins + 1)

# Call the histogram function from numpy. We receive the histogram and the bin edges
hist, bin_edges = np.histogram(d, bins=bins, density=False)

# The bin centers correspond here to the lower edges of the bins, the last higher edge we discard.
bin_center = bin_edges[:-1]
```

```python
cm = 0.393701

fig = plt.figure(figsize=(18, 10))
plt.rcParams.update({'font.size':16})
plt.bar(bin_center, hist, width = bin_size)
plt.xlabel('Value', fontsize=20)
plt.ylabel('Occurrences', fontsize=20)
plt.show()
```

## Generate Data - Sum of Multiple Dice

```python
# Sum of m dice, roled n times

n = 1000 

# Array that will hold the results
s = np.zeros(n, dtype=int)

# Number of dices rolled and added
m = 10

for i in range(0,m):
    s = s + np.random.randint(1, 7, n)
```

## Plot Histogram

```python
# Number of bins and bin size
n_bins = m * 6

bin_size = 1

bins = np.linspace(1, m * 6 + 1, n_bins + 1)

hist, bin_edges = np.histogram(s, bins=bins, density=False)
bins = bin_edges[:-1]
```

```python
mean = np.sum(s)/n
var = np.sum((s-mean)**2)/(n-1)
# Calculate the Gaussian probability density distribution based on the mean and the variance
gauss = n / (np.sqrt(2 * np.pi * var)) * np.exp(-(bins - mean)**2 / (2 * var))
```

```python
cm = 0.393701

fig = plt.figure(figsize=(18, 10))
plt.rcParams.update({'font.size':16})
plt.bar(bins, hist, width = 0.9*bin_size)
plt.plot(bins, gauss, 'ro')
plt.xlabel('Value', fontsize=20)
plt.ylabel('Occurrences', fontsize=20)
plt.show()
```
