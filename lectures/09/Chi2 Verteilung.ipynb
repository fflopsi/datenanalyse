# Initialize

## Load Libraries

```python
import random
import numpy as np
from scipy.stats import chisquare
from scipy.stats import chi2
import matplotlib.pyplot as plt
```

## Code

```python
X2_min = 0
X2_max = 20
n_X2 = 101

X2 = np.linspace(X2_min, X2_max, n_X2)

f_X2_1 = chi2.pdf(X2, 1)
f_X2_2 = chi2.pdf(X2, 2)
f_X2_4 = chi2.pdf(X2, 4)
f_X2_8 = chi2.pdf(X2, 8)

```

```python
cm = 0.393701

plt.rcParams.update({'font.size': 16})
plt.rcParams.update({'axes.labelsize': 16})

fig = plt.figure(figsize=(18*cm, 13*cm))
ax1 = fig.add_subplot(1, 1, 1)

# Plot data
#ax1.plot(X2, f_X2_1, label=r'$n=1$')
ax1.plot(X2, f_X2_2, label=r'$n=2$')
ax1.plot(X2, f_X2_4, label=r'$n=4$')
ax1.plot(X2, f_X2_8, label=r'$n=8$')

# Labels and legend
ax1.set_xlabel(r'$\chi^2$', fontsize=20)
ax1.set_ylabel(r'$f_n (\chi^2)$', fontsize=20)
ax1.legend(fontsize=20)
fig.savefig(r'Chi2.png', dpi=400, transparent=True)
plt.show()
```
